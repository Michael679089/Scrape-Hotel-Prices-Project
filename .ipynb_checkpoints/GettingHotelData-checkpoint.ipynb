````python
from datetime import date, timedelta
import re
import time
import requests
import pandas as pd

# SETTINGS
rapidapi_host = "sky-scrapper.p.rapidapi.com"
rapidapi_key = "f3068fa39amsha256be6eec9ae0ep135038jsnbf15340e581c"

needed_columns = [
    "name-of-hotel",
    "source",
    "exact-location",
    "longtitude",
    "latitude",
    "region",
    "area-type",
    "room-type",
    "price",
    "queen-or-king-bed-included",
    "pool-access-included",
    "breakfast-included",
    "wifi-included"
]

def replace_space_with_percent20(string_price):
    return string_price.replace(" ", "%20")

def convert_usd_to_php(usd):
    exchange_rate = 57.95  # Example exchange rate, you can update this with the current rate
    return usd * exchange_rate

def GetEntityID(string_hotel_name, timeout=10):
    print("GetEntityID was called")
    query = replace_space_with_percent20(str(string_hotel_name))
    url = f"https://sky-scrapper.p.rapidapi.com/api/v1/hotels/searchDestinationOrHotel?query={query}"
    headers = {
        "x-rapidapi-host": rapidapi_host,
        "x-rapidapi-key": rapidapi_key
    }
    max_retries = 5
    for attempt in range(max_retries):
        try:
            response = requests.get(url, headers=headers, timeout=timeout)
            
            if (response.status_code == 200):
                
                response_json = response.json()
                
                if (response_json["data"] is not None and len(response_json["data"]) > 0):
                    print("Data exists, now check if there are matching hotels.")
            
                    print("Checking if it's located within Philippines")
                    areThereAnyLocationsWithPhilippines = False
                    counter_index = 0
                    for data in response_json["data"]:
                        hierarchy = response_json["data"][counter_index]["hierarchy"]
                        if "philippines" in hierarchy.lower():
                            areThereAnyLocationsWithPhilippines = True
                            break
                        counter_index += 1
                    if (areThereAnyLocationsWithPhilippines == False):
                        return False
                         
                    print(f"Displaying {counter_index}th Result:")
                    print(response_json["data"][counter_index])
                    entityName = response_json["data"][counter_index]["entityName"]
                    print("\n")
                    percentage = get_similarity_percentage(string_hotel_name, entityName)
                    print("entityName:", entityName)
                    print("Similarity percentage:", percentage)
                    if (percentage >= 10):
                        print("result was above 10% ✅")
                        entityId = response_json["data"][counter_index]["entityId"]
                        return entityId
                    else:
                        print("No matching hotels found based on similarity percentage.")
                        return False
                else:
                    print("No data returned or no matching hotels found.")
                    return False
            else:
                raise Exception ("Error status code")
        except Exception as e:
            print("GetEntityID: Error: timeout. RETRY FETCHING GetEntityID")
        
def QueryHotelAmenities(entityId, timeout=10):
    print("QueryHotelAmenities was called")
    url = f"https://sky-scrapper.p.rapidapi.com/api/v1/hotels/getHotelDetails?hotelId={entityId}&entityId={entityId}&currency=PHP&market=en-US&countryCode=PH"
    headers = {
        "x-rapidapi-host": rapidapi_host,
        "x-rapidapi-key": rapidapi_key
    }
    max_retries = 3
    for attempt in range(max_retries):
        try:
            response = requests.get(url, headers=headers, timeout=timeout)
    
            if (response.status_code == 200):
                response_json = response.json()
                if (response_json.get("data") ):
                    if (response_json["data"]["amenities"]):
                        print("Successfully returning hotelAmenities")
                        print("Hotel amenities found")
                        hotelAmenities = response_json["data"]["amenities"]
                        return hotelAmenities
                    else:
                        raise Exception("No hotel amenities found")    
                else:
                    print("QueryHotelAmenities: ERROR: data is null")
                    return False
            else:
                raise Exception("Status Code Error:", response.status_code)
        except Exception as e:
            print(f"QueryHotelAmenities Error: {entityId}: {e}")
            print("retrying fetching QueryHotelAmenities")

def QueryHotelRating_And_NumOfReviews(entityId, timeout=10):
    print("QueryHotelRating_And_NumOfReviews was called")
    url = f"https://sky-scrapper.p.rapidapi.com/api/v1/hotels/getHotelDetails?hotelId={entityId}&entityId={entityId}&currency=PHP&market=en-US&countryCode=PH"
    headers = {
        "x-rapidapi-host": rapidapi_host,
        "x-rapidapi-key": rapidapi_key
    }
    max_retries = 3
    for attempt in range(max_retries):
        try:
            response = requests.get(url, headers=headers, timeout=timeout)
    
            if (response.status_code == 200):
                response_json = response.json()
                if (response_json["data"]):
                    print("data exists")
                    rating = 0.0
                    numOfReviews = 0
                    if (response_json["data"]["reviewRatingSummary"]):
                        if (response_json["data"]["reviewRatingSummary"]["score"]):
                            rating = float(response_json["data"]["reviewRatingSummary"]["score"])
                        if (response_json["data"]["reviewRatingSummary"]["countNumber"]):
                            numOfReviews = response_json["data"]["reviewRatingSummary"]["countNumber"]
                        return {
                            "rating": rating,
                            "numOfReviews": numOfReviews
                        }
                    else:
                        print("QueryHotelRating_And_NumOfReviews: reviewRatingSummary data doesn't exist")
                        break
                else:
                    raise Exception("No hotel amenities found")    
            else:
                raise Exception("Status Code Error:", response.status_code)
        except Exception as e:
            print(f"QueryHotelRating_And_NumOfReviews Error: {entityId}: {e}")
            print("retrying fetching QueryHotelRating_And_NumOfReviews")
        
def QueryHotelRoomRates(entityId, checkin_date, checkout_date, timeout=10):
    print("QueryHotelRoomRates called with entityId:", entityId, "checkin_date:", checkin_date, "checkout_date:", checkout_date)
    url = f"https://sky-scrapper.p.rapidapi.com/api/v1/hotels/getHotelPrices?hotelId={entityId}&entityId={entityId}&checkin={checkin_date}&checkout={checkout_date}&adults=1&rooms=1&currency=PHP&market=en-US&countryCode=PH"
    headers = {
        "x-rapidapi-host": rapidapi_host,
        "x-rapidapi-key": rapidapi_key
    }
    max_retries = 3
    for attempt in range(max_retries):
        try:
            # It's good practice to specify (connect_timeout, read_timeout)
            response = requests.get(url, headers=headers, timeout=timeout)
    
            if response.status_code == 200:
                response_json = response.json()
                # Use .get() to avoid KeyErrors if the API structure changes
                if ( response_json.get("data") ):
                    data = response_json.get("data")
                    if data and data.get("metaInfo"):
                        return data["metaInfo"].get("rates")
                    else:
                        # If the data is missing, retrying usually won't help
                        print(f"Data missing for {entityId}")
                        return False 
                else:
                    print("data is None")
                    return False
            else:
                print(f"Attempt {attempt + 1} failed with status: {response.status_code}")
                
        except (requests.exceptions.Timeout, requests.exceptions.ConnectionError) as e:
            print(f"QueryHotelRoomRates: Attempt {attempt + 1} timed out or connection error: {e}")
            # We don't return here; we let the loop continue to the next attempt

    # If we get here, all retries failed
    print(f"QueryHotelRoomRates: Failed to fetch rates for {entityId} after {max_retries} attempts.")
    return False

def get_similarity_percentage(text_a, text_b, ignore_words=None):
    if ignore_words is None:
        # Define common words that don't help identify a specific property
        ignore_words = {"one", "bedroom", "suite", "apartment", "hotel", "magical", "room"}

    def get_id_words(text):
        # Extract words, lowercase them, and remove ignored/common terms
        words = re.findall(r'\b\w+\b', text.lower())
        return {w for w in words if w not in ignore_words}

    set_a = get_id_words(text_a)
    set_b = get_id_words(text_b)

    # Calculate Intersection (shared words) and Union (total unique words)
    intersection = set_a.intersection(set_b)
    union = set_a.union(set_b)

    if not union:
        return 0.0

    # Jaccard Similarity Formula
    percentage = (len(intersection) / len(union)) * 100
    return round(percentage, 2)



# --------------------
# --- MAIN RUNNING ---
# --------------------

print("--- MAIN RUNNING ---")
print("getting input_df")
input_df = pd.read_excel("input-data/google (2).xlsx")

print("creating the filtered_df")
filtered_df = input_df[input_df["price"].isna()]
print("\n")

import json # This handles the string directly and adds syntax highlighting
import random

def main_process(idx, counter):
    # Setting the value just like a dictionary update
    hotelName = filtered_df.at[idx, "name-of-hotel"]
    print(f"{counter} - {hotelName}")
    print("Query the GetEntityID")

    myDict = {}

    def priceAndRoomTypeNull():
        myDict["price"] = 0.0
        myDict["room-type"] = "non-existent"      

    try: 
        entityId = GetEntityID(hotelName)
        print("THE ENTITY ID:", entityId)

        if entityId is not False and entityId is not None:
            checkin = str(date.today())
            checkout = str(date.today() + timedelta(days=1))
            print("Query RoomRates")
            hotelRoomRates = QueryHotelRoomRates(entityId, checkin, checkout)
            print("HotelRoomRates Display:")
            pretty_json = json.dumps(hotelRoomRates, indent=4)
            print(pretty_json)

            if hotelRoomRates is False:
                print("false - no hotel room rates found")
                priceAndRoomTypeNull()
                # I'm not error-ing this because maybe the hotelRoomRates are not active.
                # HotelAmenities are still online.
            else:
                # Set the values of the filtered_df
                maxAmount = len(hotelRoomRates) - 1
                chosenRandIndex = random.randint(0, maxAmount)
                myDict["room-type"] = hotelRoomRates[chosenRandIndex]["roomType"]
                if ("queen" in hotelRoomRates[chosenRandIndex]["roomType"].lower()): # (queen-or-king-bed-included)
                    myDict["queen-or-king-bed-included"] = 1
                elif ("family" in hotelRoomRates[chosenRandIndex]["roomType"].lower()): 
                    myDict["queen-or-king-bed-included"] = 1
                else:
                    myDict["queen-or-king-bed-included"] = 0
                myDict["price"] = hotelRoomRates[chosenRandIndex]["rawPrice"]
                if "Breakfast included" in hotelRoomRates[chosenRandIndex]["rateBriefFeatures"]:
                     myDict["breakfast-included"] = 1
                else:
                    myDict["breakfast-included"] = 0
            print("\n")

            print("Query Amenities")
            queryAmenities = QueryHotelAmenities(entityId)
            if (queryAmenities != False):
                queryAmenitiesContent = queryAmenities["contentV2"]
                myDict["wifi-included"] = 0
                myDict["pool-access-included"] = 0
                for amenity in queryAmenitiesContent: # (wifi-included && pool-access-included)
                    if (amenity["id"] == "WiFiAndInternet"):
                        print("set wifi-included to true")
                        myDict["wifi-included"] = 1
                    if (amenity["id"] == "PopularAndEssential"):
                        for popularEssential in amenity["items"]:
                            if (popularEssential["id"] == "Pool"):
                                print("set pool-access-included to true")
                                myDict["pool-access-included"] = 1
            else:
                print("QueryAmenities:", queryAmenities)
            print("\n")
    
            print("Querying Ratings")
            ratingAndReviews = QueryHotelRating_And_NumOfReviews(entityId)
            print("RatingAndReviewsDisplay:", ratingAndReviews)
            if (ratingAndReviews is not None):
                if (ratingAndReviews["rating"] != 0.0 and ratingAndReviews["numOfReviews"] != 0):
                    print("set rating to", ratingAndReviews["rating"])
                    myDict["rating"] = ratingAndReviews["rating"]
                    if (filtered_df.at[idx, "num-of-reviews"] < ratingAndReviews["numOfReviews"] or filtered_df.at[idx, "num-of-reviews"] is None):
                        print("set num-of-reviews to", ratingAndReviews["numOfReviews"])
                        myDict["num-of-reviews"] = ratingAndReviews["numOfReviews"]
            else:
                print("rating and reviews is None")
                if (filtered_df.at[idx, "rating"] is None):
                    print("set rating since rating is None to", 0)
                    myDict["rating"] = 0
                if (filtered_df.at[idx, "num-of-reviews"] is None):
                    print("set num-of-reviews since numOfReviews is None to", 0)
                    myDict["num-of-reviews"] = 0
            print("\n")
            
        else:
            raise Exception ("false - no match found. Skipping Row.")

        # output
        return myDict
            
    except Exception as e:
        print("main: ERROR:", e)
        priceAndRoomTypeNull()
            

    print("-----", "\n")
````
